```python
# 3.2 ZERO-SHOT MATH BASELINE

import json

with open("out/math_baseline.jsonl") as f:
    lines = f.readlines()
    metrics = json.loads(lines[0])
    results = [json.loads(line) for line in lines[1:] if line.strip()]

print(metrics)

f1a1 = []
f1a0 = []
f0a0 = []

for result in results:
    if (
        result["rewards"]["format_reward"] == 1
        and result["rewards"]["answer_reward"] == 1
    ):
        f1a1.append(result)
    elif result["rewards"]["format_reward"] == 1:
        f1a0.append(result)
    elif result["rewards"]["answer_reward"] == 0:
        f0a0.append(result)

print(f"Format 1, Answer 1: {len(f1a1)}")
print(f"Format 1, Answer 0: {len(f1a0)}")
print(f"Format 0, Answer 0: {len(f0a0)}")

print("-" * 100)

print("10 cases in which format reward is 0:")

for i in range(10):
    ex = f0a0[i]
    print(f"Completion: {ex['completion']}")
    print(f"Ground truth: {ex['ground_truth']}")
    print("\n")

print("-" * 100)
print("10 cases in which format reward is 1, but answer reward is 0:")

for i in range(10):
    ex = f1a0[i]
    print(f"Completion: {ex['completion']}")
    print(f"Ground truth: {ex['ground_truth']}")
    print("\n")
```
